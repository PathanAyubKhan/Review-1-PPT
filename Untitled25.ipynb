```python
import pandas as pd

# Load the datasets
sales_train_validation_df = pd.read_csv('sales_train_validation.csv')
calendar_df = pd.read_csv('calendar.csv')
sell_prices_df = pd.read_csv('sell_prices.csv')

# Display the first few rows of each DataFrame to understand their structure
print("Sales Train Validation DataFrame:")
display(sales_train_validation_df.head())

print("\nCalendar DataFrame:")
display(calendar_df.head())

print("\nSell Prices DataFrame:")
display(sell_prices_df.head())

# Note: Merging these datasets directly is complex due to the structure (sales data is in wide format).
# We will address merging in subsequent steps as we process the data for modeling.
```

```python
# Melt the sales data to long format
sales_data_melted = sales_train_validation_df.melt(
    id_vars=['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'],
    var_name='d',
    value_name='sales'
)

# Merge with calendar data
merged_df = pd.merge(sales_data_melted, calendar_df, on='d', how='left')

# Merge with sell prices data
merged_df = pd.merge(merged_df, sell_prices_df, on=['store_id', 'item_id', 'wm_yr_wk'], how='left')

print("\nMerged DataFrame:")
display(merged_df.head())
```

## Feature engineering

### Subtask:
Create relevant features for forecasting, such as lag features, rolling window statistics, and features related to calendar events.


**Reasoning**:
Create lag and rolling window features, extract time-based features, create dummy variables for categorical features, and handle missing values to prepare the data for forecasting.



```python
# 1. Create lag features for 'sales'
for i in range(1, 8):
    merged_df[f'sales_lag_{i}'] = merged_df.groupby(['id'])['sales'].shift(i)

# 2. Calculate rolling window statistics for 'sales'
merged_df['sales_rolling_mean_7'] = merged_df.groupby(['id'])['sales'].rolling(window=7).mean().reset_index(drop=True)
merged_df['sales_rolling_std_7'] = merged_df.groupby(['id'])['sales'].rolling(window=7).std().reset_index(drop=True)
merged_df['sales_rolling_mean_30'] = merged_df.groupby(['id'])['sales'].rolling(window=30).mean().reset_index(drop=True)
merged_df['sales_rolling_std_30'] = merged_df.groupby(['id'])['sales'].rolling(window=30).std().reset_index(drop=True)


# 3. Extract time-based features from 'date'
merged_df['year'] = merged_df['date'].dt.year
merged_df['month'] = merged_df['date'].dt.month
merged_df['dayofweek'] = merged_df['date'].dt.dayofweek
merged_df['dayofyear'] = merged_df['date'].dt.dayofyear

# 4. Create dummy variables for categorical features
categorical_cols = ['store_id', 'cat_id', 'dept_id', 'event_name_1', 'event_type_1', 'event_name_2', 'event_type_2']
merged_df = pd.get_dummies(merged_df, columns=categorical_cols, dummy_na=False)

# 5. Handle missing values introduced by feature engineering
# Fill NaN values created by lag and rolling features with 0
merged_df.fillna(0, inplace=True)

display(merged_df.head())
```

## Model selection and training

### Subtask:
Choose appropriate forecasting models (e.g., ARIMA, Prophet, or machine learning models) and train them on the historical sales data.


**Reasoning**:
Prepare the data for modeling by splitting it into training and validation sets based on time, define the features (X) and target variable (y), and train a LightGBM model.



```python
from sklearn.model_selection import train_test_split
import lightgbm as lgb

# Define features (X) and target (y)
features = [col for col in merged_df.columns if col not in ['id', 'd', 'sales', 'date', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id', 'wm_yr_wk', 'weekday', 'wday', 'month', 'year', 'event_name_1', 'event_type_1', 'event_name_2', 'event_type_2']]
target = 'sales'

X = merged_df[features]
y = merged_df[target]

# Split data into training and validation sets based on time
# Use the last 28 days for validation, as per the M5 competition guidelines
train_end_date = merged_df['date'].max() - pd.Timedelta(days=28)
X_train = X[merged_df['date'] <= train_end_date]
y_train = y[merged_df['date'] <= train_end_date]
X_val = X[merged_df['date'] > train_end_date]
y_val = y[merged_df['date'] > train_end_date]

# Initialize and train the LightGBM Regressor model
lgbm = lgb.LGBMRegressor(objective='regression',
                         n_estimators=1000,
                         learning_rate=0.05,
                         num_leaves=31,
                         n_jobs=-1,
                         seed=42)

lgbm.fit(X_train, y_train,
         eval_set=[(X_val, y_val)],
         eval_metric='rmse',
         callbacks=[lgb.early_stopping(stopping_rounds=50)])

print("LightGBM model training complete.")
```

## Forecasting

### Subtask:
Generate demand forecasts for the future period.


**Reasoning**:
Use the trained LightGBM model to generate predictions on the validation set and store these predictions in the validation DataFrame.



```python
# Make predictions on the validation set
X_val['forecasted_sales'] = lgbm.predict(X_val)

# Ensure predictions are non-negative
X_val['forecasted_sales'] = X_val['forecasted_sales'].apply(lambda x: max(0, x))

# Display the first few rows of X_val with the new forecasted sales column
display(X_val.head())
```

## Dynamic pricing engine development

### Subtask:
Develop a dynamic pricing engine that uses the demand forecasts to recommend optimal prices to maximize revenue. This might involve optimization techniques or rule-based systems.


**Reasoning**:
Define the calculate_revenue function and iterate through potential prices for each row in X_val to calculate potential revenue and find the optimal price.



```python
def calculate_revenue(forecasted_sales, sell_price):
    """Calculates total revenue."""
    return forecasted_sales * sell_price

# Iterate through each row in X_val
optimal_prices = []
max_revenues = []

# Define a range of potential price multipliers to explore
price_multipliers = [0.8, 0.9, 1.0, 1.1, 1.2] # Example: 20% below to 20% above current price

for index, row in X_val.iterrows():
    current_sell_price = row['sell_price']
    best_revenue = -1
    best_price = current_sell_price

    # Explore potential prices around the current sell price
    for multiplier in price_multipliers:
        potential_price = current_sell_price * multiplier

        # Create a temporary row for prediction with the potential price
        temp_row = row.copy()
        temp_row['sell_price'] = potential_price

        # Reshape the temporary row for prediction
        temp_row_reshaped = temp_row.drop('forecasted_sales').values.reshape(1, -1)

        # Predict demand for the potential price
        # Ensure the order of features is the same as during training
        potential_demand = lgbm.predict(temp_row_reshaped)[0]
        potential_demand = max(0, potential_demand) # Ensure non-negative demand

        # Calculate potential revenue
        potential_revenue = calculate_revenue(potential_demand, potential_price)

        # Check if this is the best revenue found so far for this item and day
        if potential_revenue > best_revenue:
            best_revenue = potential_revenue
            best_price = potential_price

    optimal_prices.append(best_price)
    max_revenues.append(best_revenue)

# Add the recommended price and max potential revenue to X_val
X_val['recommended_price'] = optimal_prices
X_val['max_potential_revenue'] = max_revenues

display(X_val.head())
```
